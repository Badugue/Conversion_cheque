```python
from PIL import Image
import os

# Répertoire source contenant les images
source_dir = 'E:/SUPDEVINCI/hackathon/support cheques/images NB/ChqNB-Flou_Lot1/'
# Répertoire de sortie
output_base_dir = 'E:/SUPDEVINCI/hackathon/support cheques/output/'

# Assurez-vous que le répertoire de sortie existe
os.makedirs(output_base_dir, exist_ok=True)

# Parcourir chaque fichier dans le répertoire source
for filename in os.listdir(source_dir):
    if filename.endswith('.jpg') or filename.endswith('.tiff'):
        image_path = os.path.join(source_dir, filename)
        image = Image.open(image_path)

        # Obtenir les dimensions de l'image
        width, height = image.size

        # Calculer les dimensions pour la division
        width_72 = int(0.72 * width)
        width_28 = width - width_72

        height_83 = int(0.83 * height)
        height_17 = height - height_83

        # Diviser l'image en deux parties verticales (72:28)
        image1 = image.crop((0, 0, width_72, height))
        image2 = image.crop((width_72, 0, width, height))

        # Diviser la première partie en deux parties horizontales (83:17)
        image3 = image1.crop((0, 0, width_72, height_83))
        image4 = image1.crop((0, height_83, width_72, height))

        # Créer un dossier de sortie pour chaque image
        output_dir = os.path.join(output_base_dir, os.path.splitext(filename)[0])
        os.makedirs(output_dir, exist_ok=True)

        # Sauvegarder les images résultantes
        image2.save(os.path.join(output_dir, 'image2.jpg'))
        image3.save(os.path.join(output_dir, 'image3.jpg'))
        image4.save(os.path.join(output_dir, 'image4.jpg'))

        print(f"Images sauvegardées dans le dossier : {output_dir}")
```
